# CHALLENGE

```python
# Import SampleData.csv to Google Colab
import pandas as pd
#from google.colab import files
#uploaded = files.upload()

# Readthe contents of SampleData.csv
df = pd.read_csv('SampleData.csv')
print("These are the contents of SampleData.csv")
print(tabulate(df, headers='keys', tablefmt='psql', showindex = True))
# Show all top 10 records
print("\n\nThese are the top 10 contents of SampleData.csv")
print(tabulate(df.head(10), headers='keys', tablefmt='psql', showindex = True))

# Show all bottom 10 records
print("\n\nThese are the bottom 10 contents of SampleData.csv")
print(tabulate(df.tail(10), headers='keys', tablefmt='psql', showindex = True))

# Eliminate all duplicated records
print("\n\nThese are the contents of SampleDataNew.csv")
df.drop_duplicates(inplace = True)
print(tabulate(df, headers='keys', tablefmt='psql', showindex = True))

# Save the contents (removed duplicate records) into new file SampleDataNew.csv
df.to_csv('SampleDataNew.csv', index=True)
```
