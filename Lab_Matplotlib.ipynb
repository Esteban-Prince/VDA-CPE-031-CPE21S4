# **===================================================**

**Line Plot**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**Scatter Plot**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]
plt.scatter(x,y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**Histogram**

```python
import matplotlib.pyplot as plt

data = [1,2,2,3,3,3,4]
plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Bar Chart**

```python
import matplotlib.pyplot as plt

categories = ['A,','B','C']
values = [5,7,3]
plt.bar(categories,values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

**Box Plot**

```python
import matplotlib.pyplot as plt

data = [[1.5]*10 + [2]*10 + [3]*10]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.xlabel("Data")
plt.ylabel("Values")
plt.show()
```

**Pie Chart**

```python
import matplotlib.pyplot as plt

labels = ['A','B','C']
sizes = [40,30,30]

plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```
