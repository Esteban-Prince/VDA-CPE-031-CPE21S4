```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = {
    'X': [0.50,1.00,1.30,1.60,2.00,2.20,2.50,3.10,3.90,4.40],
    'Y': [1.90,1.50,1.20,1.00,0.80,0.78,0.65,0.46,0.30,0.23],

}
df = pd.DataFrame(data)
df
```

```python
x = np.array([0.50,1.00,1.30,1.60,2.00,2.20,2.50,3.10,3.90,4.40])
y = np.array([1.90,1.50,1.20,1.00,0.80,0.78,0.65,0.46,0.30,0.23])

slope, intercept = np.polyfit(x, y, 1)

y_pred = slope * x + intercept

plt.scatter(x, y, label='Data Points')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()

print("Slope:", slope)
print("Intercept:", intercept)
```

According to the Data visualize, X is directly correlated to Y by the idea "As X increases, Y decreases." Therefore while X continues to increase, proportionally, Y will decrease. Since the data points are close to the regression line, this proves that Y is closely correlated to X. Furthermore, Since the regression line points towards the bottom left, it implieas that the correlation is "opposite", therefore proving the idea "While X increases, Y decreases."
