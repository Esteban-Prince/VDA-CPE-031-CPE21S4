```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("heart_2020_cleaned.csv")
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart: BMI vs Age
sns.barplot(x='39', y='Good', data=df)
plt.title('BMI vs Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.xticks(rotation=45)
plt.show()
```

This Visualization shows the relation of Ages and their Health. According to this Visualization Plot, Ages 18-24 and 35-39 are surprisingly people with the worst health than the others.

```python
# Scatter plot: BMI vs Age with Health status
plt.figure(figsize=(10, 6))
sns.scatterplot(x='39', y='8.0', hue='Good', style='American Indian/Alaskan Native', data=df)
plt.title('BMI vs Age with Health Status')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='Health Status')
plt.show()
```

This Visualization shows the scatter plot of People and their health, legends are used to distinguish the different labels. This Visualization can be interpreted to determine the health status and BMI of different people along with their races and ages.
