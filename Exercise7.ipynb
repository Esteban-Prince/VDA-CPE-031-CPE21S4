```python
import pandas as pd

# Sample DataFrame
data = {
    'User ID': [1, 2, 3, 4, 5],
    'Device Model': ['Google Pixel 5', 'OnePlus 9', 'Xiaomi Mi 11', 'Google Pixel 5', 'Iphone 12'],
    'OS': ['Android', 'Android', 'Android', 'Android', 'IOS'],
    'App Usage Time (min/day)': [393, 268, 154, 239, 187],
    'Screen on Time (hours/day)': [6.4, 4.7, 4.0, 4.8, 4.3],
    'Battery Drain (mAh/day)': [1872, 1331, 761, 1676, 1367]
}

df = pd.DataFrame(data)

# Selecting a single column
single_column = df['Device Model']
print("Single Column:\n", single_column)

# Selecting multiple columns
multiple_columns = df[['User ID', 'OS', 'Battery Drain (mAh/day)']]
print("\nMultiple Columns:\n", multiple_columns)
```

**Selecting dataframe rows**

```python
import pandas as pd

# Sample DataFrame
data = {
    'User ID': [1, 2, 3, 4, 5],
    'Device Model': ['Google Pixel 5', 'OnePlus 9', 'Xiaomi Mi 11', 'Google Pixel 5', 'Iphone 12'],
    'OS': ['Android', 'Android', 'Android', 'Android', 'IOS'],
    'App Usage Time (min/day)': [393, 268, 154, 239, 187],
    'Screen on Time (hours/day)': [6.4, 4.7, 4.0, 4.8, 4.3],
    'Battery Drain (mAh/day)': [1872, 1331, 761, 1676, 1367]
}

df = pd.DataFrame(data)

# Selecting a single row using Python indexing
single_row_python = df[2:3]
print("\nSingle row using Python indexing:\n", single_row_python)

# Selecting multiple rows using Python indexing
multiple_rows_python = df[1:4]
print("\nMultiple rows using Python indexing:\n", multiple_rows_python)
```
